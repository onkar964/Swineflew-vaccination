```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
vc = pd.read_csv("h1n1_vaccine_prediction.csv")
vc = vc.drop('unique_id', axis = 1)
```

```python
vc.isnull().sum()
```

```python
vc.h1n1_worry = vc.h1n1_worry.fillna(0.0)
vc.h1n1_awareness = vc.h1n1_awareness.fillna(1.0)
vc.antiviral_medication = vc.antiviral_medication.fillna(0.0)
vc.contact_avoidance = vc.contact_avoidance.fillna(1.0)
vc.bought_face_mask = vc.bought_face_mask.fillna(0.0)
vc.wash_hands_frequently = vc.wash_hands_frequently.fillna(1.0)
vc.avoid_large_gatherings = vc.avoid_large_gatherings.fillna(0.0)
vc.avoid_touch_face = vc.avoid_touch_face.fillna(1.0)
vc.dr_recc_h1n1_vacc = vc.dr_recc_h1n1_vacc.fillna(0.0)
vc.dr_recc_seasonal_vacc = vc.dr_recc_seasonal_vacc.fillna(0.0)
vc.chronic_medic_condition = vc.chronic_medic_condition.fillna(1.0)
vc.cont_child_undr_6_mnths = vc.cont_child_undr_6_mnths.fillna(0.0)
vc.is_health_worker = vc.is_health_worker.fillna(0.0)
vc.has_health_insur = vc.has_health_insur.fillna(0.0)
vc.is_h1n1_vacc_effective = vc.is_h1n1_vacc_effective.fillna(4.0)
vc.is_h1n1_risky = vc.is_h1n1_risky.fillna(2.0)
vc.sick_from_h1n1_vacc = vc.sick_from_h1n1_vacc.fillna(2.0)
vc.is_seas_vacc_effective = vc.is_seas_vacc_effective.fillna(4.0)
vc.is_seas_risky = vc.is_seas_risky.fillna(2.0)
vc.sick_from_seas_vacc = vc.sick_from_seas_vacc.fillna(1.0)
vc.qualification = vc.qualification.fillna('College Graduate')
vc.income_level = vc.income_level.fillna('<= $75,000, Above Poverty')
vc.marital_status = vc.marital_status.fillna('Married')
vc.housing_status = vc.housing_status.fillna('Own')
vc.employment = vc.employment.fillna('Employed')
vc.no_of_adults = vc.no_of_adults.fillna(1.0)
vc.no_of_children = vc.no_of_children.fillna(1.0)
vc.reduced_outside_home_cont = vc.reduced_outside_home_cont.fillna(0.0)
```

```python
vc.isnull().sum()
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
```

```python
vc[vc.select_dtypes(include=['object']).columns] = vc[vc.select_dtypes(include=['object']).columns].apply(le.fit_transform)
```

```python
vc.head()
```

```python
vc.shape
```

```python
train_x = vc.iloc[:,0:-1]
train_y = vc.iloc[:,-1]
```

```python
train_x1 = train_x
```

```python
train_x = np.array(train_x)
train_y = np.array(train_y)
```

```python
from boruta import BorutaPy

from sklearn.ensemble import RandomForestClassifier
rf = RandomForestClassifier()

boruta_feature_selection = BorutaPy(rf, verbose=2)

boruta_feature_selection.fit(train_x, train_y)
```

```python
df1 = pd.DataFrame({"Feature_nae" : train_x1.columns, "Importance" : boruta_feature_selection.support_})
```

```python
from sklearn.feature_selection import SelectKBest
from sklearn.feature_selection import chi2
```

```python
test = SelectKBest(score_func= chi2, k='all')
fitted_cr = test.fit(train_x, train_y)
```

```python
df2 = pd.DataFrame({"Feature_nae" : train_x.columns, "Importance" : list(fitted_cr.scores_)})
df2
```

```python
df2.sort_values("Importance", ascending=False)
```
